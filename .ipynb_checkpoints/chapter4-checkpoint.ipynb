## 4.1 Parallel coordinate plot
Unfortunately, `seaborn` does not provide function to draw a parallel coordinate plot directly. However, `pandas` allows us to draw it. Morever, `plotly` makes it possible for us to draw an interactive parallel coordinate plot.
### 4.1.1 Draw parallel coordinate plot via `pandas`
Click [here](https://pandas.pydata.org/docs/reference/api/pandas.plotting.parallel_coordinates.html?highlight=parallel_coordinates) to see the documentation of the `parallel-coordinates` function provided by `pandas.plotting`.

```python
import pandas
import seaborn as sns
from pandas.plotting import parallel_coordinates
```

```python
df = sns.load_dataset('iris')
df.info()
```

By specify parameter `class_column`, we can distinguish the different classes of data flows in the parallel coordinate plot. 

There are a few differences between `parallel_coordinates()` in `pandas` and `ggparcoord()` in `GGally`:
1. `class_column` is a required parameter in `parallel_coordinates()`, which means that if we want to use `pandas` to draw parallel coordinate plots, we have to give flows in different classes different colors.
2. `parallel_coordinates()` cannot scale your data, if you plan to use some scalers, you have to scale it before draw a plot.
3. `parallel_coordinates()` cannot apply spines to deal with repeating data. However, we could specify parameter `alpha` to solve this problem.

```python
parallel_coordinates(frame=df,
                    class_column='species',
                    alpha=0.3)
```

### 4.1.2 Draw interactive parallel coordinate plot via `plotly`
`plotly.express` provides a function also called `parallel_coordinates()`, which allows us to draw interactive parallel coordinates plots. Click [here](https://plotly.com/python/parallel-coordinates-plot/) to see the detailed tutorial of 

```python
import plotly.express as px
```

```python
df = px.data.iris()
fig = px.parallel_coordinates(df, color="species_id", labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
```

## 4.2 Bioplot
Unfortunately, neither `seaborn` nor `matplotlib` provides a direct function to draw a bioplot. However, we can use `Python` to draw it step by step (Although it is not efficient).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bioplot(reduced_data, variable, pc_variances, cov):
    """plot bioplots for the firts two principle components

    : param reduced_data: data after PCA
    : param pc: all the principle components
    : param variable: the name of the variables of the data set
    : param pc_variances: the proportion of variance of the principe components
    : param cov: covariance matrix of the original data 
    """

    c = np.linalg.eig(cov)
    pc = c[1]
    x = reduced_data[:, 0]
    y = reduced_data[:, 1]
    n = pc.shape[1]

    scalex = 1.0/(x.max() - x.min())
    scaley = 1.0/(y.max() - y.min())

    plt.scatter(x*scalex, y*scaley)

    

    for i in range(n):
        plt.arrow(0, 0, pc[0,i], pc[1,i], alpha=0.5)
        plt.text(pc[0,i]*1.1, pc[1,i]*1.1, variable[i], ha='center', va='center')
    
    plt.xlabel("PC1({})".format(pc_variances[0]))
    plt.xlabel("PC1({})".format(pc_variances[1]))
    plt.title("Compositional biplot")
    plt.grid()
    plt.show()
    
```

```python
from sklearn.decomposition import PCA 
```

```python
df.columns
```

```python
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width',]].values
pca = PCA(n_components=2)
new_X = pca.fit_transform(X)
```

```python
bioplot(new_X, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width',],
        pca.explained_variance_ratio_, pca.get_covariance())
```

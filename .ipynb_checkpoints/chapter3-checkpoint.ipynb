# Chapter 3: Two continuous variables

In this chapter, we will focus on how to plot the relationship between two continuous variables with the help of `seaborn`. Basically, we have two choices: scatter plot and heatmap.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
```

## 3.1 Basics and implications

`seaborn.scatterplot` can plot the scatter plot with two variables conveniently

```python
df = sns.load_dataset("iris")
df.info()
```

```python
sns.scatterplot(data=df, x="sepal_length", y="sepal_width")
```

The result shows that there is no obvious linear relationship between "sepal_length" and "sepal_width".

## 3.1.2 Overlaping

When dealing with a huge number of data points, it is unavoidable that the data points will overlap with some others, which might influence the visualization result of the scatter plot. In that case, we usually have three choices:
1. Reduce the transparency of data points.
2. Use open circles to represent data points.
3. Use smaller circles to represent data points.

```python
diamonds= sns.load_dataset("diamonds")
diamonds.info()
```

```python
sns.scatterplot(data=diamonds, x="x", y="carat")
```

According to the result, the default setting of `seaborn.scatterplot()` can somewhat decrease the negative effect of overlapping, because the data points become open circles when overlapping happens.

Now we try to reduce the transparency of the data point.

```python
sns.scatterplot(data=diamonds, x="x", y="carat", alpha=0.3, edgecolors="blue")
```

Then we try to use smaller data points.

```python
sns.scatterplot(data=diamonds, x="x", y="carat", size=0.1, edgecolors="blue")
```

### 3.1.3 Scatter plot with extra information
When specify the parameter `hue`, we can add extra information to the scatter plot. Let us go back to the `iris` dataset, according to the scatter plot of "sepal_length" and "sepal_width", we can hardly find useful information. 

However, if we specify the parameter `hue="species"`

```python
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
```

We can discover that there are differences among different species' data points.

### 3.1.4 Scatter plot matrices
When we want to explore the relationships between several variable pairs at the same time, scatter plot matrices is a good choice. In order to do that, we should apply `seaborn.PairGrid`.

```python
# Make the PairGrid
g = sns.PairGrid(diamonds,
                 x_vars=["carat", "x", "y", "z"], y_vars=["carat", "x", "y", "z"],)

# Draw a dot plot using the scatterplot function
g.map(sns.scatterplot)
```

## 3.2 Plot the joint distribution of two continuous variables

Instead of `scatterplot`, `seaborn` provides a more efficient function `displot` to show the joint distribution of two continuous variables. We have two options for parameter `kind` when drawing the joint distribution plot of two continuous variables: 

1. `kde`: Draw the contour lines plot
2. `hist`: Draw the heatmap

```python
sns.displot(data=df, x="sepal_length", y="petal_width", kind="kde")
```

```python
sns.displot(data=df, x="sepal_length", y="petal_width", kind="hist")
```

By specify the parameter `rug` to be `True`, we can add density informations to the axises of the two variables.

```python
sns.displot(data=df, x="sepal_length", y="petal_width", kind="kde", rug=True)
```

```python
sns.displot(data=df, x="sepal_length", y="petal_width", kind="hist", rug=True)
```

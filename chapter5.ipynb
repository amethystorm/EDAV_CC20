# Chapter 5 Multidimentional Categorical variable

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 5.1 Barchart

### 5.1.1 Stacked bar chart

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="species", hue='sex',multiple='stack')
```

### 5.1.2 Grouped bar chart

```python
sns.barplot(data=penguins, x="species",y='body_mass_g',hue='sex')
```

### 5.1.3 Grouped bar chart with facet

```python
bar_plot_facet = sns.FacetGrid(penguins, row='sex', hue='sex')
bar_plot_facet.map(sns.barplot,'species','body_mass_g')
```

### 5.1.4 Grouped bar chart with three categorical variables

```python
sns.catplot(data=penguins, x="species", y="body_mass_g", col="sex", hue='island', kind="bar", height=4, aspect=.6,)
```

## 5.2 Mosaic plots

Seaborn does not have original support for mosaic plot. We use statsmodels.graphics instead

```python
from statsmodels.graphics.mosaicplot import mosaic
```

```python
mosaic(data=penguins,index=['species','sex'])
```

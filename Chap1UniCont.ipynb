# Chapter 1 Unidimentional continuous variable

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1.1 Histogram

### 1.1.1 Basics

We use sns.histplot to create a histogram. The "data" variable is to specify the target dataframe. The 'x' variable is the name  of the target value.

```python
titanic = sns.load_dataset('titanic')
sns.histplot(data=titanic, x ="age")
```

### 1.1.2 Different types of histogram

The default histogram gives us the count of value in each bin on y-scale. We can modify this using 'stat' variable.

```python
sns.histplot(data=titanic, x ="age")
```

The following histogram give us frequency (the number of observations divided by the bin width) on y axis. You can also try stat = 'probability', 'percent' and 'density'.

```python
sns.histplot(data=titanic, x ="age", stat='frequency')
```

The following histogram gives us cumulative count.

```python
sns.histplot(data=titanic, x ="age", stat='count', cumulative=True)
```

### 1.1.3 Adjust bins

sns.histplot provides 3 methods to adjust bins: bins, binwidth, binrange.

```python
sns.histplot(data=titanic, x ="age", bins=5)
```

```python
sns.histplot(data=titanic, x ="age", binwidth=5)
```

The binrange variable sets the lowest and highest bin edge.

```python
sns.histplot(data=titanic, x ="age", binrange=(40,60))
```

## 1.2 Boxplot

### 1.2.1 Basics

We use sns.boxplot to create boxplot. The following code gives us a single boxplot.

```python
sns.boxplot(data=titanic, x='age')
```

### 1.2.2 Multiple boxplots

To have multiple boxplots, we need to specify y axis.

```python
sns.boxplot(data=titanic, x='age', y='sex')
```

## 1.3 Ridgeline plot

### 1.3.1 Basics

The seaborn package doesn't support ridgeline plot. However, we can create it manually. First we create a facet grid. Then we use the map function to create a kernel density function in each grid.

```python
ridgeline_plot = sns.FacetGrid(titanic, row='class', hue='class', height=1, aspect=5)
ridgeline_plot.map(sns.kdeplot,'age')
```
